```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
df = pd.read_csv('data/beta_simulator_windows/My_data/driving_log.csv', header=None)
```

```python
df.columns = ['center', 'left', 'right', 'steering_angle', 'throttle', 'break', 'speed']
df.head()
```

```python
df['steering_angle'].plot()
```

```python
img = Image.open(df['center'][0])
plt.imshow(img)
```

```python
np.asarray(img).shape
```

```python
def load_image(fname):
    img = Image.open(fname)
    return np.asarray(img)

def process_image(img):
    #160, 320, 3 => 60,80,3
    return img[10:130:2, ::4, :]
```

```python
img = process_image(load_image(df['center'][0]))
plt.imshow(img)
```

```python
X = [process_image(load_image(fname)) for fname in df['center']]
X += [process_image(load_image(fname)) for fname in df['left']]
X += [process_image(load_image(fname)) for fname in df['right']]

y = np.array(df['steering_angle'])
y = np.concatenate([y,y+0.4, y-0.4])
```

```python
X = np.array(X)
X.shape
```

```python
plt.plot(y)
```
