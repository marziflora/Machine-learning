```python
def preprocesing(sentence):
    return [word for word in simple_preprocess(sentence) if word not in STOPWORDS]

def read_sequence(filename):
    with open(filename, 'rb') as f:
        for line in f:
            yield preprocesing(line)
```

TOPIC MODELING - LDA

```python
sentences_light = np.random.permutation(sentence)
sentences_light = sentences_light[:1000]
```

```python
%time dictionary = Dictionary(sentences_light)
```

```python
list(dictionary.items())[:10] #every word has its ID
```

```python
bow_corpus = [dictionary.doc2bow(sent) for sent in sentences_light]
```

```python
dictionary.doc2bow(['car']) #Token ID:14,  occurence: 1 time
```

```python
dictionary.doc2bow(['hotel'])
```

```python
dictionary.token2id['hotel'] #
```

```python
lda_model = LdaMulticore(bow_corpus, id2word=dictionary, num_topics=100, passes=20, workers=8)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} /nWords: {}'.format(idx,topic))
```

```python
lda_vis = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dictionary)
pyLDAvis.display(lda_vis)
#disadvantage: slow time processing, vulnerable for polish words
```
